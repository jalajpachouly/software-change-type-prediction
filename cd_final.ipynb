```python
# -*- coding: utf-8 -*-
"""cd_final_with_cross_validation_corrected.py

This script performs multiclass text classification on software defect reports.
It uses SMOTE to balance the dataset and evaluates the performance of various
machine learning models on balanced data, using k=10 cross-fold validation.

It generates the following plots:
1. Confusion Matrix for each traditional ML model (Balanced Data)
2. ROC Curve for each traditional ML model (Balanced Data)
3. Confusion Matrix for BERT (Balanced Data)
4. ROC Curve for BERT (Balanced Data)

All other plots related to class distribution, word clouds, and general evaluation metrics (box/bar plots)
for both balanced and unbalanced datasets have been removed as per the request.
Processing for unbalanced datasets has also been skipped.

Additionally, this script now saves the true labels, predicted labels, and
prediction probabilities for ALL models (traditional ML and BERT) into
separate JSON files for further detailed analysis.
"""

# Install necessary libraries
!pip install transformers -q
!pip install datasets -q
!pip install wordcloud -q
# Ensure scikit-learn is up-to-date
!pip install scikit-learn --upgrade
# Ensure imbalanced-learn is compatible with the new scikit-learn version
!pip install imbalanced-learn --upgrade -q # For SMOTE


# Import libraries
import pandas as pd
import numpy as np # Ensure numpy is imported
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import nltk
import torch
import json # Changed from pickle to json

# NLTK downloads
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Download the punkt_tab resource

# Transformers and tokenizers
from transformers import BertTokenizer, BertForSequenceClassification
from transformers.optimization import get_linear_schedule_with_warmup
from torch.optim import AdamW # Use standard PyTorch AdamW

# Data handling
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import LabelEncoder

# Evaluation metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_fscore_support, roc_curve, auc, RocCurveDisplay, ConfusionMatrixDisplay

# Machine Learning models
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier

# Handling class imbalance
from imblearn.over_sampling import SMOTE

# Handle warnings
import warnings
warnings.filterwarnings('ignore')

# For progress bar
from tqdm.notebook import tqdm

# Load the dataset
data = pd.read_csv('dataset.csv')

# Display first few rows (kept for informational purposes, no plot)
print("First few rows of the dataset:")
print(data.head())

# Check for missing values (kept for informational purposes, no plot)
print("\nMissing values in each column:")
print(data.isnull().sum())

# Distribution of labels (kept for informational purposes, no plot)
print("\nClass distribution:")
print(data['target'].value_counts())

# Original Class Distribution Plot - REMOVED
# plt.figure(figsize=(10,6))
# sns.countplot(x='target', data=data, order=data['target'].value_counts().index)
# plt.xticks(rotation=45)
# plt.title('Original Class Distribution')
# plt.tight_layout()
# plt.savefig('original_class_distribution.png')
# plt.show()

import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize
    words = nltk.word_tokenize(text)
    # Remove stopwords and lemmatize
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    # Join back into text
    return ' '.join(words)

# Apply preprocessing
tqdm.pandas()
data['cleaned_text'] = data['report'].progress_apply(preprocess_text)

# Display cleaned text (kept for informational purposes, no plot)
print("\nSample of original and cleaned text:")
print(data[['report', 'cleaned_text']].head())

# Encode labels
label_encoder = LabelEncoder()
data['label_encoded'] = label_encoder.fit_transform(data['target'])
num_classes = len(label_encoder.classes_)
class_names = label_encoder.classes_

# Generate word clouds for each class - REMOVED
def generate_wordclouds(data):
    classes = data['target'].unique()
    for cls in classes:
        text = ' '.join(data[data['target'] == cls]['cleaned_text'])
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
        # plt.figure(figsize=(10, 5))
        # plt.imshow(wordcloud, interpolation='bilinear')
        # plt.axis('off')
        # plt.title(f'Word Cloud for {cls}')
        # plt.tight_layout()
        # plt.savefig(f'wordcloud_{cls}.png')
        # plt.close()

generate_wordclouds(data) # Call the function, but it won't save/show plots now

# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Linear SVM': LinearSVC(random_state=42),
    'Multinomial NB': MultinomialNB(),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}

# Function to plot Confusion Matrix
def plot_confusion_matrix_heatmap(y_true, y_pred, labels, title, filename):
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    fig, ax = plt.subplots(figsize=(10, 8))
    disp.plot(cmap=plt.cm.Blues, ax=ax)
    plt.title(title)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()

# Function to plot ROC curve (multi-class)
def plot_roc_curve(y_true, y_proba, num_classes, class_names, title, filename):
    plt.figure(figsize=(10, 8))
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(num_classes):
        # Ensure y_true is binary for roc_curve for each class
        binary_y_true = (np.array(y_true) == i).astype(int)
        fpr[i], tpr[i], _ = roc_curve(binary_y_true, y_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    for i in range(num_classes):
        plt.plot(fpr[i], tpr[i], label=f'ROC curve of class {class_names[i]} (area = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', label='Chance level (AUC = 0.5)')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()


# Function to perform cross-validation and collect metrics
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

def perform_cv_metrics(models, X, y, data_desc, balance=False, skip_unbalanced=False, num_classes=None, class_names=None):
    if skip_unbalanced and not balance:
        print(f"Skipping {data_desc} data run as per script configuration.")
        return pd.DataFrame()

    # Initialize metrics dictionary (kept for internal calculation, not for plotting general metrics)
    metrics_dict = {'Model': list(models.keys()), 'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-Score': []}
    # For storing metrics per model across folds
    per_model_metrics = {model_name: {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-Score': [], 'y_true': [], 'y_pred': [], 'y_proba': []} for model_name in models.keys()}

    # Initialize StratifiedKFold
    kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    fold = 1
    last_y_train = None  # To store y_train from the last fold
    for train_index, test_index in kf.split(X, y):
        print(f"\nFold {fold} for {data_desc} data")
        X_train, X_val = X.iloc[train_index], X.iloc[test_index]
        y_train, y_val = y.iloc[train_index], y.iloc[test_index]

        # Vectorize text using TF-IDF
        tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))

        X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
        X_val_tfidf = tfidf_vectorizer.transform(X_val)

        if balance:
            # Apply SMOTE to balance the training data
            smote = SMOTE(random_state=42)
            X_train_tfidf, y_train_smote = smote.fit_resample(X_train_tfidf, y_train)
        else:
            y_train_smote = y_train  # Keep original labels if not balancing

        # Store y_train from the last fold for balancing verification
        if fold == 10 and balance:
            last_y_train = y_train_smote.copy()

        # Train and evaluate models
        for model_name, model in models.items():
            # Clone the model to avoid reusing the same model across folds
            model_clone = clone(model)
            model_clone.fit(X_train_tfidf, y_train_smote)
            y_pred = model_clone.predict(X_val_tfidf)

            y_proba = None
            if hasattr(model_clone, "predict_proba"):
                y_proba = model_clone.predict_proba(X_val_tfidf)
            elif model_name == 'Linear SVM':
                y_proba = model_clone.decision_function(X_val_tfidf)
                # For multiclass, decision_function returns (n_samples, n_classes).
                # If it's a binary case and returns 1D, you might need to convert it.
                # Here, we assume multiclass behavior.
                # If it returns decision values for OvR, these can be used as scores.
                # For proper probabilities, CalibratedClassifierCV might be needed.


            accuracy = accuracy_score(y_val, y_pred)
            precision, recall, f1_score, _ = precision_recall_fscore_support(y_val, y_pred, average='weighted', zero_division=0)

            per_model_metrics[model_name]['Accuracy'].append(accuracy)
            per_model_metrics[model_name]['Precision'].append(precision)
            per_model_metrics[model_name]['Recall'].append(recall)
            per_model_metrics[model_name]['F1-Score'].append(f1_score)
            per_model_metrics[model_name]['y_true'].extend(y_val.tolist())
            per_model_metrics[model_name]['y_pred'].extend(y_pred.tolist())
            if y_proba is not None:
                per_model_metrics[model_name]['y_proba'].extend(y_proba.tolist())

            print(f"{model_name} - Accuracy: {accuracy:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}, F1-Score: {f1_score:.2f}")

        fold += 1

    # Calculate average metrics for each model (for internal reporting, not plotting)
    for model_name in models.keys():
        metrics_dict['Accuracy'].append(np.mean(per_model_metrics[model_name]['Accuracy']))
        metrics_dict['Precision'].append(np.mean(per_model_metrics[model_name]['Precision']))
        metrics_dict['Recall'].append(np.mean(per_model_metrics[model_name]['Recall']))
        metrics_dict['F1-Score'].append(np.mean(per_model_metrics[model_name]['F1-Score']))

        # Plot Confusion Matrix and ROC curve if balanced and for the entire dataset after CV
        if balance and num_classes is not None and class_names is not None:
            # Aggregate true and predicted values across all folds for overall CM and ROC
            all_y_true = per_model_metrics[model_name]['y_true']
            all_y_pred = per_model_metrics[model_name]['y_pred']
            all_y_proba = np.array(per_model_metrics[model_name]['y_proba'])

            if all_y_true and all_y_pred:
                plot_confusion_matrix_heatmap(all_y_true, all_y_pred,
                                              labels=list(range(num_classes)),
                                              title=f'Confusion Matrix - {model_name} (Balanced)',
                                              filename=f'confusion_matrix_{model_name.replace(" ", "_").lower()}_balanced.png')

            # Only plot ROC if probabilities are available and correctly shaped for multi-class
            # For LinearSVC, ensure that decision_function output is correctly handled for multi-class ROC
            if all_y_proba.shape and all_y_proba.ndim == 2 and all_y_proba.shape[1] == num_classes:
                plot_roc_curve(all_y_true, all_y_proba, num_classes, class_names,
                               f'ROC Curve - {model_name} (Balanced)',
                               f'roc_curve_{model_name.replace(" ", "_").lower()}_balanced.png')
            else:
                print(f"Skipping ROC curve for {model_name} due to missing or incorrectly shaped probabilities. "
                      f"Expected shape: (n_samples, {num_classes}), Actual shape: {all_y_proba.shape}")

            # --- Save Traditional ML Model Outputs to a JSON File ---
            print(f"\nSaving {model_name} true labels, predicted labels, and probabilities to '{model_name.replace(' ', '_').lower()}_outputs.json'...")
            data_to_save_ml = {
                "y_true": np.array(all_y_true).tolist(),
                "y_pred": np.array(all_y_pred).tolist(),
                "y_proba": (np.array(all_y_proba).tolist() if all_y_proba.shape else []) # Handle cases where y_proba might be empty
            }

            with open(f"{model_name.replace(' ', '_').lower()}_outputs.json", "w") as f:
                json.dump(data_to_save_ml, f)
            print(f"{model_name} outputs saved successfully to '{model_name.replace(' ', '_').lower()}_outputs.json'.")
            # --- End of Saving Traditional ML Model Outputs ---


    # Plot training label distribution after SMOTE (Balanced) for the last fold - REMOVED
    # if balance and last_y_train is not None:
    #     print("\nAfter SMOTE in the last fold, label distribution:")
    #     unique, counts = np.unique(last_y_train, return_counts=True)
    #     print(dict(zip(label_encoder.inverse_transform(unique), counts)))
    #     plt.figure(figsize=(12, 5))
    #     sns.countplot(x=last_y_train, palette='viridis')
    #     plt.title('Training Data After SMOTE (Balanced) - Last Fold')
    #     plt.xlabel('Classes')
    #     plt.ylabel('Count')
    #     plt.xticks(ticks=range(len(label_encoder.classes_)), labels=label_encoder.classes_, rotation=45)
    #     plt.tight_layout()
    #     plt.savefig('training_data_after_smote_last_fold.png')
    #     plt.show()

    return pd.DataFrame(metrics_dict)

# Prepare data
X = data['cleaned_text']
y = data['label_encoded']

# Collect metrics for unbalanced data using cross-validation - SKIPPED
metrics_unbalanced = perform_cv_metrics(models, X, y, data_desc='unbalanced', balance=False, skip_unbalanced=True)

# Collect metrics for balanced data using cross-validation
metrics_balanced = perform_cv_metrics(models, X, y, data_desc='balanced', balance=True, num_classes=num_classes, class_names=class_names)

# Visualization of Evaluation Metrics - REMOVED
def plot_metrics_box(metrics_df, title, filename):
    if metrics_df.empty:
        print(f"Skipping box plot for {title} as no data is available.")
        return
    # plt.figure(figsize=(10,6))
    # metrics_melted = metrics_df.melt(id_vars='Model', var_name='Metric', value_name='Score')
    # sns.boxplot(x='Metric', y='Score', data=metrics_melted, palette='Set3')
    # plt.title(f'Evaluation Metrics - {title} Data', fontsize=14)
    # plt.ylabel('Score', fontsize=12)
    # plt.xlabel('Metrics', fontsize=12)
    # plt.xticks(fontsize=10)
    # plt.yticks(fontsize=10)
    # plt.tight_layout()
    # plt.savefig(filename)
    # plt.show()

def plot_metrics_bar(metrics_df, title, filename):
    if metrics_df.empty:
        print(f"Skipping bar plot for {title} as no data is available.")
        return
    # metrics_melted = metrics_df.melt(id_vars='Model', var_name='Metric', value_name='Score')
    # plt.figure(figsize=(10,6))
    # sns.barplot(x='Metric', y='Score', hue='Model', data=metrics_melted, palette='Set2')
    # plt.title(f'Evaluation Metrics - {title} Data', fontsize=14)
    # plt.ylabel('Score', fontsize=12)
    # plt.xlabel('Metrics', fontsize=12)
    # plt.xticks(fontsize=10)
    # plt.yticks(fontsize=10)
    # plt.legend(title='Model', fontsize=10)
    # plt.tight_layout()
    # plt.savefig(filename)
    # plt.show()

# Generate plots for unbalanced data - SKIPPED
# plot_metrics_box(metrics_unbalanced, 'Unbalanced', 'eval_metrics_box_unbalanced.png')
# plot_metrics_bar(metrics_unbalanced, 'Unbalanced', 'eval_metrics_bar_unbalanced.png')

# Generate plots for balanced data - REMOVED
# plot_metrics_box(metrics_balanced, 'Balanced', 'eval_metrics_box_balanced.png')
# plot_metrics_bar(metrics_balanced, 'Balanced', 'eval_metrics_bar_balanced.png')

# ------- BERT Section -------

# For BERT, we will use the original train/test split

# Split data for BERT models
X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(
    data['cleaned_text'], data['label_encoded'], test_size=0.2, random_state=42, stratify=data['label_encoded'])

# Combine training data for BERT
train_data_bert = pd.DataFrame({
    'cleaned_text': X_train_ml,
    'label': y_train_ml
})

# Oversample the minority classes
from sklearn.utils import resample

train_data_list = [train_data_bert[train_data_bert['label'] == i] for i in range(len(label_encoder.classes_))]
max_size = max(len(d) for d in train_data_list)

train_data_balanced = pd.concat([
    resample(d, replace=True, n_samples=max_size, random_state=42) if len(d) < max_size else d
    for d in train_data_list
])

# Shuffle the balanced training data
train_data_balanced = train_data_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Plot BERT training label distribution after oversampling (Balanced) - REMOVED
# plt.figure(figsize=(12, 5))
# sns.countplot(x='label', data=train_data_balanced, palette='magma')
# plt.title('BERT Training Data After Oversampling (Balanced)')
# plt.xlabel('Classes')
# plt.ylabel('Count')
# plt.xticks(ticks=range(len(label_encoder.classes_)), labels=label_encoder.classes_, rotation=45)
# plt.tight_layout()
# plt.savefig('bert_training_data_after_oversampling.png')
# plt.show()

# Tokenize data
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_function(texts):
    return tokenizer(
        texts, add_special_tokens=True, max_length=128, padding='max_length',
        truncation=True, return_attention_mask=True, return_tensors='pt')

# Balanced data for BERT
X_train_balanced = train_data_balanced['cleaned_text']
y_train_balanced = train_data_balanced['label']

# Tokenize data for BERT
print("\nTokenizing data for BERT...")

# Balanced data
train_encodings_balanced = tokenize_function(X_train_balanced.tolist())

test_encodings = tokenize_function(X_test_ml.tolist())

# Convert labels to tensors
train_labels_balanced = torch.tensor(y_train_balanced.tolist())
test_labels = torch.tensor(y_test_ml.tolist())

from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler

# Function to create data loaders
def create_data_loader(encodings, labels, batch_size=16):
    dataset = TensorDataset(
        encodings['input_ids'], encodings['attention_mask'], labels)
    data_loader = DataLoader(dataset, sampler=RandomSampler(dataset), batch_size=batch_size)
    return data_loader

# Create data loaders
train_dataloader_balanced = create_data_loader(train_encodings_balanced, train_labels_balanced)
test_dataset = TensorDataset(
    test_encodings['input_ids'], test_encodings['attention_mask'], test_labels)
test_dataloader = DataLoader(test_dataset, sampler=SequentialSampler(test_dataset), batch_size=16)


# Function to train and evaluate BERT model
def train_evaluate_bert(train_dataloader, data_desc, num_classes, class_names):
    # Check if GPU is available
    device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
    print(f"\nUsing device: {device} for {data_desc} data")

    # Load pre-trained BERT model
    model = BertForSequenceClassification.from_pretrained(
        'bert-base-uncased', num_labels=num_classes)
    model.to(device)

    # Optimizer and Scheduler
    optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8)
    epochs = 4
    total_steps = len(train_dataloader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer, num_warmup_steps=0, num_training_steps=total_steps)

    # Training function
    def train():
        model.train()
        total_loss = 0
        progress_bar = tqdm(train_dataloader, desc="Training")
        for batch in progress_bar:
            optimizer.zero_grad()
            input_ids = batch[0].to(device)
            attention_mask = batch[1].to(device)
            labels = batch[2].to(device)
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
            loss.backward()
            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
        avg_loss = total_loss / len(train_dataloader)
        print(f"\nAverage Training Loss: {avg_loss:.4f}")

    # Evaluation function
    def evaluate():
        model.eval()
        predictions, true_labels, all_logits = [], [], []
        progress_bar = tqdm(test_dataloader, desc="Evaluating")
        with torch.no_grad():
            for batch in progress_bar:
                input_ids = batch[0].to(device)
                attention_mask = batch[1].to(device)
                labels = batch[2].to('cpu').numpy()
                outputs = model(input_ids, attention_mask=attention_mask)
                logits = outputs.logits
                preds = torch.argmax(logits, axis=1).detach().cpu().numpy()
                predictions.extend(preds)
                true_labels.extend(labels)
                all_logits.extend(logits.detach().cpu().numpy())

        accuracy = accuracy_score(true_labels, predictions)
        report = classification_report(true_labels, predictions, target_names=class_names, output_dict=True)
        print(f"Accuracy on {data_desc} data: {accuracy:.2f}")
        print("Classification Report:")
        print(classification_report(true_labels, predictions, target_names=class_names))

        # Plot Confusion Matrix for BERT
        plot_confusion_matrix_heatmap(true_labels, predictions, labels=list(range(num_classes)),
                                      title=f'Confusion Matrix - BERT ({data_desc.capitalize()})',
                                      filename=f'confusion_matrix_bert_{data_desc.lower()}.png')

        # Plot ROC Curve for BERT (using softmax probabilities from logits)
        softmax_logits = torch.nn.functional.softmax(torch.tensor(all_logits), dim=1).numpy()
        plot_roc_curve(true_labels, softmax_logits, num_classes, class_names,
                       title=f'ROC Curve - BERT ({data_desc.capitalize()})',
                       filename=f'roc_curve_bert_{data_desc.lower()}.png')


        return accuracy, report, true_labels, predictions, softmax_logits

    # Lists to store metrics (kept for internal reporting, not plotting)
    accuracies = []
    reports = []
    # These will be populated and returned at the end of training loop
    final_y_true, final_y_pred, final_y_proba = [], [], []

    # Training loop
    for epoch in range(epochs):
        print(f"\nEpoch {epoch+1}/{epochs} for {data_desc} data")
        train()
        acc, report, yt, yp, ypb = evaluate()
        accuracies.append(acc)
        reports.append(report)
        # For BERT, we are interested in the final evaluation's true, pred, proba
        # If you need per-epoch, you'd handle that differently.
        # Here, we return the last epoch's evaluation results.
        final_y_true = yt
        final_y_pred = yp
        final_y_proba = ypb

    return accuracies, reports, final_y_true, final_y_pred, final_y_proba

# Train and evaluate BERT on balanced data only
bert_balanced_accuracies, bert_balanced_reports, bert_balanced_y_true, bert_balanced_y_pred, bert_balanced_y_proba = train_evaluate_bert(train_dataloader_balanced, "balanced", num_classes, class_names)

# --- Saving BERT Outputs to a JSON File ---
print("\nSaving BERT true labels, predicted labels, and probabilities to 'bert_outputs.json'...")
# Convert to plain Python types (no internal NumPy types)
data_to_save_bert = {
    "y_true": np.array(bert_balanced_y_true).tolist(),
    "y_pred": np.array(bert_balanced_y_pred).tolist(),
    "y_proba": np.array(bert_balanced_y_proba).tolist()
}

with open("bert_outputs.json", "w") as f:
    json.dump(data_to_save_bert, f)
print("BERT outputs saved successfully to 'bert_outputs.json'.")
# --- End of Saving BERT Outputs ---

def add_bert_metrics(metrics_df, bert_reports, bert_accuracies, data_desc):
    # Take the last epoch's metrics
    bert_report = bert_reports[-1]
    # Create a DataFrame for BERT metrics
    bert_metrics_df = pd.DataFrame([{
        'Model': 'BERT',
        'Accuracy': bert_accuracies[-1],
        'Precision': bert_report['weighted avg']['precision'],
        'Recall': bert_report['weighted avg']['recall'],
        'F1-Score': bert_report['weighted avg']['f1-score']
    }])
    # Concatenate with the existing metrics_df
    metrics_df = pd.concat([metrics_df, bert_metrics_df], ignore_index=True)
    return metrics_df

# metrics_unbalanced = add_bert_metrics(metrics_unbalanced, bert_unbalanced_reports, bert_unbalanced_accuracies, "unbalanced") # SKIPPED
metrics_balanced = add_bert_metrics(metrics_balanced, bert_balanced_reports, bert_balanced_accuracies, "balanced")

# Update plots with BERT metrics (only for balanced) - REMOVED
# plot_metrics_box(metrics_balanced, 'Balanced', 'eval_metrics_box_balanced_with_bert.png')
# plot_metrics_bar(metrics_balanced, 'Balanced', 'eval_metrics_bar_balanced_with_bert.png')

print("\nAll tasks completed successfully.")
```
